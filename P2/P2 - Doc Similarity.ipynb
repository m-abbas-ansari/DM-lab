```python
import pandas as pd
import numpy as np
import glob
import re
```

```python
docs = []

for doc in glob.glob("docs/*"):
    with open(doc, "r") as f:
        docs.append(re.findall(r'\w+', f.read()))
```

```python
vocab = set()
for doc in docs:
    for word in doc:
        if word not in vocab:
            vocab.add(word)
            
vocab = list(vocab)
```

```python
frequency_matrix = {'Words': vocab}

for i, doc in enumerate(docs):
    count_matrix = {word: 0 for word in vocab}
    for word in doc:
        count_matrix[word] += 1
    frequency_matrix['D'+str(i)] = list(count_matrix.values())
```

```python
freq_table = pd.DataFrame(frequency_matrix)
```

```python
freq_table
```

```python
count_values = []
for i, doc in enumerate(freq_table):
    if i == 0:
        continue
    count_values.append(list(freq_table[doc]))
```

```python
len(count_values)
```

```python
def compute_cosine_sim(dat1, dat2):
    assert len(dat1) == len(dat2)
    
    mag1 = np.linalg.norm(dat1)
    mag2 = np.linalg.norm(dat2)
    score = np.dot(dat1, dat2)/(mag1*mag2)
    return score
```

```python
similarity_matrix = {'Doc1': [], 'Doc2': [], 'Cosine Similarity': []}

for i in range(len(count_values)):
    doc1 = count_values[i]
    for j in range(i+1, len(count_values)):
        doc2 = count_values[j]
        
        cos_sim = compute_cosine_sim(doc1[1:], doc2[1:])
        similarity_matrix['Doc1'].append('D' + str(i))
        similarity_matrix['Doc2'].append('D' + str(j))
        similarity_matrix['Cosine Similarity'].append(cos_sim)
```

```python
similarity_table = pd.DataFrame(similarity_matrix)
```
